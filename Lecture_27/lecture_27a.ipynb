**Vector norms**

<font size = "4">

- We often want to compute $\mathbf{y} = f(\mathbf{x})$ where $\mathbf{x} \in \mathbb{R}^n$ and $\mathbf{y}\in \mathbb{R}^m$

- We can define the error vector

$$\mathbf{e} \coloneqq \hat{\mathbf{x}} - \mathbf{x}$$

- But we need a replacement for the absolute value in order to talk about its magnitude.

- A **norm** $||\cdot||$ is a function satisfying

    - $\|\mathbf{x}\| \geq 0$ and $\|\mathbf{x}\| = 0$ only when $\mathbf{x} = \mathbf{0}$

    - $\|\alpha\mathbf{x}\| = |\alpha|\cdot \|\mathbf{x}\|$ for any scalar $\alpha$

    - $\|\mathbf{x} + \mathbf{y}\| \leq \|\mathbf{x}\| + \|\mathbf{y}\|$

- The most common norms are the $p$-norms, defined for $p \geq 1$:

$$\|\mathbf{x}\|_p = \left(\sum_{i=1}^n |x_i|^p \right)^{1/p}$$

- The most common values of $p$ are:

    - $p = 1$:   $\quad \|\mathbf{x}\|_1 = \sum\limits_{i=1}^n |x_i|$

    - $p = 2$:   $\quad\|\mathbf{x}\|_2 = \left(\sum\limits_{i=1}^n |x_i|^2 \right)^{1/2} = \sqrt{\sum\limits_{i=1}^n |x_i|^2}$

    - $p = \infty$: $\quad \|\mathbf{x}\|_\infty = \lim\limits_{p\to \infty}\left(\sum\limits_{i=1}^n |x_i|^p \right)^{1/p} = \max\limits_{1\leq i\leq n}|x_i|$

**Matrix norms**

<font size = "4">

- In order to define the condition number of a problem, we will also need to define a **norm** for matrices.

- The general definition is the same. A matrix norm must satisfy:

    - $\|\mathbf{A}\| \geq 0$ and $\|\mathbf{A}\| = 0$ only when $\mathbf{A} = \mathbf{0}$

    - $\|\alpha\mathbf{A}\| = |\alpha|\cdot \|\mathbf{A}\|$ for any scalar $\alpha$

    - $\|\mathbf{A} + \mathbf{B}\| \leq \|\mathbf{A}\| + \|\mathbf{B}\|$


- The most common are those *induced* by a vector norm:

    $$\|\mathbf{A}\|_1 = \max_{\mathbf{x}\neq \mathbf{0}} \frac{\|\mathbf{A}\mathbf{x}\|_1}{\|\mathbf{x}\|_1}\qquad \|\mathbf{A}\|_2 = \max_{\mathbf{x}\neq \mathbf{0}} \frac{\|\mathbf{A}\mathbf{x}\|_2}{\|\mathbf{x}\|_2}\qquad \|\mathbf{A}\|_\infty = \max_{\mathbf{x}\neq \mathbf{0}} \frac{\|\mathbf{A}\mathbf{x}\|_\infty}{\|\mathbf{x}\|_\infty}$$


- Matrix norms that are induced by a vector norm satisfy two additional properties:

    - $\|\mathbf{A}\mathbf{B}\| \leq \|\mathbf{A}\|\cdot\|\mathbf{B}\|$

    - $\|\mathbf{A}\mathbf{x}\| \leq \|\mathbf{A}\|\cdot\|\mathbf{x}\|$


- The induced matrix norms above are defined by taking the maximum ratio of vector norms. But it can be demonstrated that for $p = 1, 2, \infty$ we can characterize them directly as for an $m\times n$ matrix as:

    - $\|\mathbf{A}\|_1 = \max\limits_{j}\sum\limits_{i=1}^m|a_{ij}|\quad$ (maximum absolute column sum)

    - $\|\mathbf{A}\|_\infty = \max\limits_{i}\sum\limits_{j=1}^n|a_{ij}|\quad$ (maximum absolute row sum)

    - $\|\mathbf{A}\|_2 = $ largest singular value of $\mathbf{A}$ = square root of largest eigenvalue of $\mathbf{A}^T\mathbf{A}$ (or $\mathbf{A}\mathbf{A}^T$)


**Example**

$$\mathbf{A} = \begin{bmatrix}2 & -1 & 1\\ 1 & 0 & 1\\ 3 & -1 & 4\end{bmatrix}


```python
import numpy as np

A = np.array([[2, -1, 1], [1, 0, 1], [3, -1, 4]])

print("1-norm of A:", np.linalg.norm(A, 1))
print("Maximum abs. column sum:", np.sum(np.abs(A),axis = 0).max())
print()

print("infinity-norm of A:", np.linalg.norm(A, np.inf))
print("Maximum abs. row sum:", np.sum(np.abs(A),axis = 1).max())

print()
print("2-norm of A:", np.linalg.norm(A, 2))
U, S, V = np.linalg.svd(A)
print("max singular value:", S.max())
eigval, eigvecs = np.linalg.eig(A.T @ A)
print("square root, max eigenval of A.T @ A:", np.sqrt(eigval.max()))
eigval, eigvecs = np.linalg.eig(A @ A.T)
print("square root, max eigenval of A @ A.T:", np.sqrt(eigval.max()))
```

**Condition Number for Vector-Valued problems**

<font size = "4">

- We want to compute $\mathbf{y} = f(\mathbf{x})$ where $\mathbf{x} \in \mathbb{R}^n$ and $\mathbf{y}\in \mathbb{R}^m$

- We compute $\hat{\mathbf{y}} = f(\hat{\mathbf{x}})$ instead (perhaps using a backward stable algorithm).

- The condition number is

\begin{align*}
    \textrm{cond}(f) = \lim_{\delta \to 0}\ \sup_{\|\hat{\mathbf{x}} - \mathbf{x}\|_{\mathbb{R}^n}\leq \delta} \dfrac{\|f(\hat{\mathbf{x}}) - f(\mathbf{x})\|_{\mathbb{R}^m}\ /\ \|f(\mathbf{x})\|_{\mathbb{R}^m} }{ \|\hat{\mathbf{x}} - \mathbf{x}\|_{\mathbb{R}^n}\ /\ \|\mathbf{x}\|_{\mathbb{R}^n}} = \lim_{\delta \to 0}\ \sup_{\|\hat{\mathbf{x}} - \mathbf{x}\|_{\mathbb{R}^n}\leq \delta} \dfrac{ \|\hat{\mathbf{y}} - \mathbf{y}\|_{\mathbb{R}^m}\ /\ \|\mathbf{y}\|_{\mathbb{R}^m}}{ \|\hat{\mathbf{x}} - \mathbf{x}\|_{\mathbb{R}^n}\ /\ \|\mathbf{x}\|_{\mathbb{R}^n}}
\end{align*}

- $\|\cdot\|_{\mathbf{R}^n}$ and $\|\cdot\|_{\mathbf{R}^m}$ are norms on the input and output space, respectively. We can choose any combination of norms between the two spaces.

- For simplicity, we will just right "$\|\cdot\|$" for an unspecified norm.

- Recall relative condition number for a differentiable scalar-valued function of 1 variable:

\begin{align*}
    \textrm{cond}(f) = \frac{|f'(x)|}{|f(x)|\ /\  |x|} = \frac{|x|\cdot |f'(x)|}{|f(x)|}
\end{align*}

- In general, the relative condition number for a differentiable function $f$ is:

\begin{align*}
    \textrm{cond}(f) = \frac{\|\mathbf{J}_f(\mathbf{x})\|}{\|f(\mathbf{x})\|\ /\  \|\mathbf{x}\|} = \frac{\|\mathbf{x}\|\cdot \|\mathbf{J}_f(\mathbf{x})\|}{\|f(\mathbf{x})\|}
\end{align*}

- Above, $\mathbf{J}_f(\mathbf{x})$ is the **Jacobian** matrix of partial derivatives of $f$:

$$\left[\mathbf{J}_f(\mathbf{x}) \right]_{ij} = \frac{\partial f_i}{\partial x_j}(\mathbf{x})$$


<font size = "4">

**Example**

- Consider computing $y = f(\mathbf{x}) = f(x_1, x_2) = x_1 - x_2$

- This is a mapping from $\mathbf{x} \in \mathbb{R}^2$ to $y \in \mathbb{R}$. We'll choose the $\infty$-norm for $\mathbb{R}^2$ (the infinity norm for $\mathbb{R}$ is just absolute value)

- Jacobian:

$$\mathbf{J}_f(\mathbf{x}) = \begin{bmatrix} \frac{\partial f}{\partial x_1}(\mathbf{x}) & \frac{\partial f}{\partial x_2}(\mathbf{x}) \end{bmatrix} = \begin{bmatrix} 1 & -1 \end{bmatrix}$$

- Condition number:

$$\textrm{cond}(f) = \frac{\|\mathbf{x}\|_\infty \cdot \| \mathbf{J}_f(\mathbf{x}) \|_\infty}{|f(\mathbf{x})|} = \frac{\max\{|x_1|, |x_2|\} \cdot 2}{|x_1 - x_2|}$$

- If $x_1 \approx x_2$, then the small denominator results in a huge condition number. So subtracting two numbers that are very close together is an ill-conditioned problem. This is known as **catastrophic cancellation**.

<font size = "4">

**Example: Catastrophic Cancellation**

- To compute the sample standard deviation of a set of numbers $x_1, \dots, x_n$ we can try one of two approaches.

- The *2-pass* formula:

    - Compute $\bar{x} = \dfrac{1}{n}\sum\limits_{i=1}^nx_i$ by looping over the values of the sequence.

    - Compute $\sigma^2 = \dfrac{1}{n-1}\sum\limits_{i=1}^n(x_i - \bar{x})^2$ in another loop.

- The *1-pass* formula:

    - Loop over the sequence once to compute $\bar{x} = \dfrac{1}{n}\sum\limits_{i=1}^nx_i$ and $s = \sum\limits_{i=1}^n x_i^2$

    - Compute $\sigma^2 = \dfrac{n}{n-1}\left(\dfrac{s}{n} - \bar{x}^2\right)$

- For the sequence $x_i = 10^8 + i$, the 1-pass algorithm is less accurate because

    $$\frac{s}{n} = \frac{1}{n}\sum_{i=1}^n(10^8 + i)^2 \approx \left(\sum_{i=1}^n(10^8 + i) \right)^2 = \bar{x}$$

- For reference, the true variance of the sequence is $\sigma^2 = \dfrac{n(n+1)}{12}$

```python
n = 10
shift = 1e8

x_seq = np.arange(1,n+1) + shift

true_var = n * (n + 1) / 12

####### 2-pass formula

x_bar = 0
for x in x_seq:
    x_bar += x
x_bar /= n

var_x = 0
for x in x_seq:
    var_x += (x - x_bar)**2
var_x /= n-1

print("Computed variance (2-pass):", var_x)
print("Relative error:", abs(var_x - true_var)/true_var)

print()

####### 1-pass formula

x_bar = 0
x_sqr = 0

for x in x_seq:
    x_sqr += x**2
    x_bar += x
x_bar /= n

var_x = n * (x_sqr/n - x_bar**2) / (n-1)

print("Computed variance (1-pass):", var_x)
print("Relative error:", abs(var_x - true_var)/true_var)
```

**Multi-dimensional problems**

For 1-dimensional problems we argued that from the definition

$$\textrm{cond}(f) = \lim_{\delta \to 0}\ \sup_{|\hat{x} - x|\leq \delta} \dfrac{|\hat{y} - y|\ /\ |y| }{ |\hat{x} - x|\ /\ |x|}$$

it was reasonable to assume for small $|\hat{x} - x|$ that

$$\textrm{cond}(f) \approx \dfrac{|\hat{y} - y|\ /\ |y| }{ |\hat{x} - x|\ /\ |x|}$$

at least for differentiable functions. This led us to:

$$ |\textrm{relative forward error}| \approx \textrm{cond}(f) \times  |\textrm{relative backward error}|$$


For problems with multi-dimensional inputs, we shouldn't ignore the "$\sup$" (maximum) in the definition, so it is much safer to say

$$ |\textrm{relative forward error}| \lesssim \textrm{cond}(f) \times  |\textrm{relative backward error}|$$

<br>

Regardless, a **backward stable algorithm applied to a well-conditioned algorithm will lead to an accurate solution**.
