```python
class Voter:
    def __init__(self, voter_id, age, county, party, income_bracket, turnout_history):
        self.voter_id = voter_id
        self.age = age
        self.county = county
        self.party = party
        self.income_bracket = income_bracket
        self.turnout_history = turnout_history  # list of 0/1

    def turnout_rate(self):
        if not self.turnout_history:
            return 0.0
        return sum(self.turnout_history) / len(self.turnout_history)

    def __repr__(self):
        return (
            f"Voter(id={self.voter_id}, age={self.age}, "
            f"county='{self.county}', party='{self.party}', "
            f"income='{self.income_bracket}')"
        )

# Example use for one data point from a generated dataset
data = generate_dataset(n=100)
single_data_pt = data[0]
turn_out = [single_data_pt["t_2016"], single_data_pt["t_2018"],
    single_data_pt["t_2020"], single_data_pt["t_2022"]]

voter = Voter(single_data_pt["voter_id"],
            single_data_pt["age"],
            single_data_pt["county"],
            single_data_pt["party"],
            single_data_pt["income_bracket"],
            turn_out)

print(voter)
print(voter.turnout_rate())
```

### Data Structure: Linked List

`VoterRegistry` stores `Voter` objects as a single-linked list of `VoterNode` objects. This design mimics the `Node` class and the linked-list traversal pattern: a `_head` pointer and a `current = current.next` loop.

`VoterNode` is a simplified version of that `Node`, here we store `Voter` objects directly.

### Sorting: Merge Sort

Sorting uses merge sort, extended with a `key` parameter (like Python's built-in `sorted`). Merge sort was chosen because:
- O(n log n) worst case: unlike quicksort, performance is consistent regardless of initial order (random, sorted, reversed).
- Stable: equal-keyed elements preserve their original relative order.

Workflow: extract all `Voter` objects to a Python list -> merge sort with the chosen key -> rebuild the linked list in sorted order.

`income_bracket` needs a custom numeric mapping (`"<50K"-> 0`, `"50-100K"-> 1`, `">100K"-> 2`) because its natural order is not alphabetical.

### Searching: Linear Scan

`search(condition)` traverses every node once, collecting 0-based indices where `condition(voter)` is `True`. This is O(n): the same asymptotic cost as any unsorted sequential structure.

### Benchmarking

Timing uses `timeit.default_timer` and log-log plots.

```python
# VoterNode:
# Stores a Voter object; the primitive-only restriction from the original Node is removed.

class VoterNode:
    """A node in the VoterRegistry linked list. Holds one Voter object."""

    def __init__(self, voter):
        self._data = voter
        self._next = None

    def get_data(self):
        return self._data
    def set_data(self, voter):
        self._data = voter
    data = property(get_data, set_data)

    def get_next(self):
        return self._next
    def set_next(self, node_next):
        self._next = node_next
    next = property(get_next, set_next)

    def __repr__(self):
        return f"VoterNode({self._data})"


# income_bracket strings do not sort in their natural order alphabetically
# ('<' < '5' < '>'), so we map each bracket to an integer rank.
INCOME_ORDER = {"<50K": 0, "50-100K": 1, ">100K": 2}

# Maps each sortable attribute name to a function that extracts the
# comparison value from a Voter object.
SORT_KEYS = {
    "voter_id":       lambda v: v.voter_id,
    "age":            lambda v: v.age,
    "county":         lambda v: v.county,
    "party":          lambda v: v.party,
    "income_bracket": lambda v: INCOME_ORDER[v.income_bracket],
    "t_2016":         lambda v: v.turnout_history[0],
    "t_2018":         lambda v: v.turnout_history[1],
    "t_2020":         lambda v: v.turnout_history[2],
    "t_2022":         lambda v: v.turnout_history[3],
    "turnout_rate":   lambda v: v.turnout_rate(),
}
```

```python
# Merge sort, extended with a 'key' parameter so any voter attribute 
# (or computed value) can drive the comparison.

def merge_sort(a_list, key=lambda x: x):
    """
    Sort a_list in-place in ascending order using merge sort.
    key: function extracting the comparison value from each element.
    Complexity: O(n log n) worst case. Stable.
    """
    if len(a_list) <= 1:
        return

    mid = len(a_list) // 2
    left_half = a_list[:mid]
    right_half = a_list[mid:]

    merge_sort(left_half, key)
    merge_sort(right_half, key)

    i, j, k = 0, 0, 0
    while i < len(left_half) and j < len(right_half):
        if key(left_half[i]) <= key(right_half[j]):
            a_list[k] = left_half[i]
            i += 1
        else:
            a_list[k] = right_half[j]
            j += 1
        k += 1

    while i < len(left_half):
        a_list[k] = left_half[i]
        i += 1
        k += 1

    while j < len(right_half):
        a_list[k] = right_half[j]
        j += 1
        k += 1
```

```python
class VoterRegistry:
    """
    Linked-list data structure for storing and managing Voter objects.

    Supports:
    - Initialization from generate_dataset() output
    - Appending new voter records
    - Sorting in ascending order by any voter attribute (or turnout_rate)
    - Searching by Boolean conditions on voter attributes
    """

    def __init__(self, data):
        """
        Build the registry from the list of dicts returned by generate_dataset().
        Each dict is converted to a Voter object and appended to the linked list.
        """
        self._head = None
        self._tail = None
        self._size = 0
        for record in data:
            self._append_voter(self._dict_to_voter(record))

    # internal helpers 

    def _dict_to_voter(self, d):
        """Convert a generate_dataset dict to a Voter object."""
        return Voter(
            d["voter_id"], d["age"], d["county"], d["party"],
            d["income_bracket"],
            [d["t_2016"], d["t_2018"], d["t_2020"], d["t_2022"]]
        )

    def _append_voter(self, voter):
        """Append a Voter object to the tail of the linked list."""
        new_node = VoterNode(voter)
        if self._head is None:
            self._head = new_node
            self._tail = new_node
        else:
            self._tail.next = new_node
            self._tail = new_node
        self._size += 1

    def _rebuild_from_list(self, voter_list):
        """Replace linked list contents with a Python list of Voter objects."""
        self._head = None
        self._tail = None
        self._size = 0
        for voter in voter_list:
            self._append_voter(voter)

    # public interface

    def to_list(self):
        """Traverse the linked list and return all Voter objects as a Python list."""
        result = []
        current = self._head
        while current is not None:
            result.append(current.data)
            current = current.next
        return result

    def append(self, new_data):
        """
        Add new voter records to the registry.
        Accepts a single dict or a list of dicts (from generate_dataset).
        The registry is not automatically re-sorted; call sort() afterwards if needed.
        """
        if isinstance(new_data, dict):
            new_data = [new_data]
        for record in new_data:
            self._append_voter(self._dict_to_voter(record))

    def sort(self, by="voter_id"):
        """
        Sort the registry in ascending order by the given attribute.
        Valid keys: voter_id, age, county, party, income_bracket,
                    t_2016, t_2018, t_2020, t_2022, turnout_rate.
        Uses merge sort: O(n log n), stable.
        """
        if by not in SORT_KEYS:
            raise ValueError(
                f"Unknown sort key '{by}'. Choose from: {list(SORT_KEYS.keys())}"
            )
        voters = self.to_list()
        merge_sort(voters, key=SORT_KEYS[by])
        self._rebuild_from_list(voters)

    def search(self, condition):
        """
        Return 0-based indices of voters satisfying condition.
        condition: callable (Voter -> bool).

        Examples:
            registry.search(lambda v: v.age > 30)
            registry.search(lambda v: v.county == "Fulton")
            registry.search(lambda v: v.party == "Democrat" and v.turnout_rate() == 1.0)
        """
        indices = []
        current = self._head
        idx = 0
        while current is not None:
            if condition(current.data):
                indices.append(idx)
            current = current.next
            idx += 1
        return indices

    def __len__(self):
        return self._size

    def __repr__(self):
        voters = self.to_list()
        n_show = min(5, self._size)
        lines = [f"VoterRegistry ({self._size} voters):"]
        for i in range(n_show):
            lines.append(f"  [{i}] {voters[i]}")
        if self._size > n_show:
            lines.append(f"  ... ({self._size - n_show} more)")
        return "\n".join(lines)
```

## Benchmarking

We time `sort()` across different dataset sizes and initial orderings, and `search()` across sizes. Timing uses `timeit.default_timer`, results are plotted on a log-log scale.

Expected behavior:
- `sort`: O(n log n). Merge sort always splits and merges every level regardless of initial order, so timing should be similar for random, pre-sorted, and reversed input.
- `search`: O(n). Linear scan —> time grows proportionally to n.

```python
from timeit import default_timer
import matplotlib.pyplot as plt
import numpy as np


def time_sort(n, order="random", by="age", repeats=5):
    """
    Time VoterRegistry.sort() on a dataset of size n.
    order: 'random', 'sorted', or 'reversed'
    Returns average time in seconds over `repeats` independent runs.
    Each run re-generates the data so the initial state is always fresh.
    """
    times = []
    for _ in range(repeats):
        reg = VoterRegistry(generate_dataset(n=n))
        if order == "sorted":
            reg.sort(by=by)
        elif order == "reversed":
            reg.sort(by=by)
            voters = reg.to_list()
            voters.reverse()
            reg._rebuild_from_list(voters)
        start = default_timer()
        reg.sort(by=by)
        times.append(default_timer() - start)
    return sum(times) / repeats


def time_search(n, repeats=5):
    """
    Time VoterRegistry.search() on a dataset of size n.
    Returns average time in seconds over `repeats` independent runs.
    """
    times = []
    for _ in range(repeats):
        reg = VoterRegistry(generate_dataset(n=n))
        start = default_timer()
        _ = reg.search(lambda v: v.age > 40)
        times.append(default_timer() - start)
    return sum(times) / repeats
```

```python
sizes = [100, 500, 1000, 2000, 5000]

print("Sort timing by dataset size (random order)")
sort_times_random = []
for n in sizes:
    t = time_sort(n, order="random")
    sort_times_random.append(t)
    print(f"  n={n:5d}: {t*1000:.3f} ms")

print()
print("Sort timing by initial ordering (all sizes)")
ordering_times = {"random": [], "sorted": [], "reversed": []}
for n in sizes:
    for order in ["random", "sorted", "reversed"]:
        t = time_sort(n, order=order)
        ordering_times[order].append(t)
    print(f"  n={n:5d}  random={ordering_times['random'][-1]*1000:.2f}ms  "
          f"sorted={ordering_times['sorted'][-1]*1000:.2f}ms  "
          f"reversed={ordering_times['reversed'][-1]*1000:.2f}ms")

print()
print("Search timing by dataset size")
search_times = []
for n in sizes:
    t = time_search(n)
    search_times.append(t)
    print(f"  n={n:5d}: {t*1000:.3f} ms")
```

```python
n_arr = np.array(sizes)
sort_arr = np.array(sort_times_random)
search_arr = np.array(search_times)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# Plot 1: Sort scaling 
ax = axes[0]
nlogn = n_arr * np.log(n_arr)
C = sort_arr[-1] / nlogn[-1]
ax.loglog(n_arr, sort_arr, 'k.--', label="Measured", markersize=8)
ax.loglog(n_arr, C * nlogn, '.-', label=r"$\mathcal{O}(n \log n)$", markersize=6)
ax.set_xlabel("n (voters)")
ax.set_ylabel("Time (s)")
ax.set_title("sort() Scaling")
ax.legend()

# Plot 2: Search scaling
ax = axes[1]
C2 = search_arr[-1] / n_arr[-1]
ax.loglog(n_arr, search_arr, 'k.--', label="Measured", markersize=8)
ax.loglog(n_arr, C2 * n_arr, '.-', label=r"$\mathcal{O}(n)$", markersize=6)
ax.set_xlabel("n (voters)")
ax.set_ylabel("Time (s)")
ax.set_title("search() Scaling")
ax.legend()

# Plot 3: Initial ordering effect
ax = axes[2]
styles = {"random": ("steelblue", "o"), "sorted": ("seagreen", "s"), "reversed": ("tomato", "^")}
for order, (color, marker) in styles.items():
    t_ms = np.array(ordering_times[order]) * 1000
    ax.plot(sizes, t_ms, marker=marker, color=color,
            label=order, linewidth=2, markersize=7)
ax.set_xlabel("n (voters)")
ax.set_ylabel("Time (ms)")
ax.set_title("sort() by Initial Ordering")
ax.legend()

plt.tight_layout()
plt.show()

```

**Key observation:** The three ordering curves (random, sorted, reversed) closely overlap across most dataset sizes. This is expected, merge sort always divides the list in half and merges every level regardless of what order the data starts in, so initial ordering does not affect the number of operations. Any visible differences in the timing are assumed to be due to system noise rather than algorithmic behavior.

This contrasts with algorithms like insertion sort, which is O(n) on already-sorted data but O(n²) on random or reversed data.
