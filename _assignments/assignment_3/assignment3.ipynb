**(Q1)** The bubble sort algorithm can be modified to "bubble" in both directions. The first pass moves "up" the list, and the maximum element "bubbles" to the top. The second pass moves "down" the list, and the minimum element "bubbles" to the bottom. This alternating pattern continues until no more passes are necessary. 

Write a function that implements this algorithm.

````python
def bidirectional_bubble_sort(lst):
    n = len(lst)
    start = 0
    end = n - 1

    while start < end:
        # Forward pass: bubble largest to the right
        exchange = False
        for i in range(start, end):
            if lst[i] > lst[i + 1]:
                lst[i], lst[i + 1] = lst[i + 1], lst[i]
                exchange = True
        end -= 1

        if not exchange:
            break

        # Backward pass: bubble smallest to the left
        exchange = False
        for i in range(end, start, -1):
            if lst[i] < lst[i - 1]:
                lst[i], lst[i - 1] = lst[i - 1], lst[i]
                exchange = True
        start += 1

        if not exchange:
            break

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
bidirectional_bubble_sort(lst)
print(lst)
````

**(Q2)** The implementation of the merge sort algorithm we saw had two deficiencies:

1. It used **slicing** as follows:

```python
        left_half = a_list[:mid]
        right_half = a_list[mid:]
        merge_sort(left_half)
        merge_sort(right_half)
```
Each slicing operation incurs an additional $O(k)$ cost, where $k$ is the size of the slice.

2. Once the slicing operation is executed, we need extra memory to store the two halves of the list independently of the original list. This increases the algorithm's *space* complexity.

Implement a version of merge sort that does not use slicing and sorts the list **in-place**. Some starter code is provided below for you to complete.

What sorting algorithm should be used for merging two sorted sub-lists?

**Insertion sort should be used to merge the two sorted sub-lists.**


````python
def merge_sort(a_list):
    merge_sort_helper(a_list, 0, len(a_list))

def merge_sort_helper(lst, start_index, end_index):
    # 1. Base case of recursion
    if end_index - start_index <= 1:
        return

    # 2. Two recursive calls to merge_sort_helper
    mid = (start_index + end_index) // 2
    merge_sort_helper(lst, start_index, mid)
    merge_sort_helper(lst, mid, end_index)

    # 3. Merge the two halfs of the list (in-place!)
    i = mid
    while i < end_index:
        cur_val = lst[i]
        j = i
        while j > start_index and lst[j - 1] > cur_val:
            lst[j] = lst[j - 1]
            j -= 1
        lst[j] = cur_val
        i += 1

a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
merge_sort(a_list)
print(a_list)
````

**(Q3)** One way to improve the quicksort algorithm is to enforce a "partition limit". Instead of treating $n = 1$ as the base case of the recursion, we choose a short length $m > 1$ for the base case. When the length of the sub-list is $\leq m$, we end the recursion and sort it directly using insertion sort.

- Implement this version of quicksort where the partition limit is included as an additional input parameter. For $m = 1$, it reduces to the original quicksort.

- Perform benchmark tests on large lists of randomly generated integers. Vary both the size of the lists ($n$) and the partition limit ($m$) and see how performance is affected, in terms of time.

````python
import random
import time

def insertion_sort_sublist(a_list, first, last):
    for i in range(first + 1, last + 1):
        cur_val = a_list[i]
        cur_pos = i
        while cur_pos > first and a_list[cur_pos - 1] > cur_val:
            a_list[cur_pos] = a_list[cur_pos - 1]
            cur_pos -= 1
        a_list[cur_pos] = cur_val

def partition(a_list, first, last):
    pivot_val = a_list[first]
    left_mark = first + 1
    right_mark = last
    done = False
    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_val:
            left_mark += 1
        while left_mark <= right_mark and a_list[right_mark] >= pivot_val:
            right_mark -= 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = a_list[right_mark], a_list[left_mark]
    a_list[first], a_list[right_mark] = a_list[right_mark], a_list[first]
    return right_mark

def quick_sort_limit(a_list, m=1):
    quick_sort_limit_helper(a_list, 0, len(a_list) - 1, m)

def quick_sort_limit_helper(a_list, first, last, m):
    if last - first + 1 <= m:
        insertion_sort_sublist(a_list, first, last)
        return
    if first < last:
        split = partition(a_list, first, last)
        quick_sort_limit_helper(a_list, first, split - 1, m)
        quick_sort_limit_helper(a_list, split + 1, last, m)

# Benchmark: 
# vary n (list size) and m (partition limit)

ns = [1000, 5000, 10000, 50000]
ms = [1, 5, 10, 20, 50]

for n in ns:
    for m in ms:
        lst = [random.randint(0, 100000) for _ in range(n)]
        start = time.time()
        quick_sort_limit(lst, m)
        elapsed = time.time() - start
        print(f"n={n:>6}, m={m:>3}: {elapsed:.6f} seconds")
````

1. Implement a variant of quicksort where the **last** element of the list is used as the pivot value.

2. Implement a variant of quicksort where the **middle** element of the list is used as the pivot value.

Last element as pivot:

````python
def partition_last(a_list, first, last):
    # Swap last element to front, then use standard partition logic
    a_list[first], a_list[last] = a_list[last], a_list[first]
    pivot_val = a_list[first]
    left_mark = first + 1
    right_mark = last
    done = False
    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_val:
            left_mark += 1
        while left_mark <= right_mark and a_list[right_mark] >= pivot_val:
            right_mark -= 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = a_list[right_mark], a_list[left_mark]
    a_list[first], a_list[right_mark] = a_list[right_mark], a_list[first]
    return right_mark

def quick_sort_last(a_list):
    quick_sort_last_helper(a_list, 0, len(a_list) - 1)

def quick_sort_last_helper(a_list, first, last):
    if first < last:
        split = partition_last(a_list, first, last)
        quick_sort_last_helper(a_list, first, split - 1)
        quick_sort_last_helper(a_list, split + 1, last)

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
quick_sort_last(lst)
print("Last-element pivot:", lst)
````

````python
def partition_middle(a_list, first, last):
    # Swap middle element to front, then use standard partition logic
    mid = (first + last) // 2
    a_list[first], a_list[mid] = a_list[mid], a_list[first]
    pivot_val = a_list[first]
    left_mark = first + 1
    right_mark = last
    done = False
    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_val:
            left_mark += 1
        while left_mark <= right_mark and a_list[right_mark] >= pivot_val:
            right_mark -= 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = a_list[right_mark], a_list[left_mark]
    a_list[first], a_list[right_mark] = a_list[right_mark], a_list[first]
    return right_mark

def quick_sort_middle(a_list):
    quick_sort_middle_helper(a_list, 0, len(a_list) - 1)

def quick_sort_middle_helper(a_list, first, last):
    if first < last:
        split = partition_middle(a_list, first, last)
        quick_sort_middle_helper(a_list, first, split - 1)
        quick_sort_middle_helper(a_list, split + 1, last)

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
quick_sort_middle(lst)
print("Middle-element pivot:", lst)
````

**(Q5)** Implement quicksort where the "median of three" method is used for selecting the pivot. That is, compare the first, middle, and last elements of a list and use the median as the pivot value.

````python
def median_of_three(a_list, first, last):
    mid = (first + last) // 2
    # Sort the three candidates so that a_list[mid] holds the median
    if a_list[first] > a_list[mid]:
        a_list[first], a_list[mid] = a_list[mid], a_list[first]
    if a_list[first] > a_list[last]:
        a_list[first], a_list[last] = a_list[last], a_list[first]
    if a_list[mid] > a_list[last]:
        a_list[mid], a_list[last] = a_list[last], a_list[mid]
    return mid  # index of the median

def partition_median3(a_list, first, last):
    median_idx = median_of_three(a_list, first, last)
    # Swap median to the front, then use standard partition logic
    a_list[first], a_list[median_idx] = a_list[median_idx], a_list[first]
    pivot_val = a_list[first]
    left_mark = first + 1
    right_mark = last
    done = False
    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_val:
            left_mark += 1
        while left_mark <= right_mark and a_list[right_mark] >= pivot_val:
            right_mark -= 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = a_list[right_mark], a_list[left_mark]
    a_list[first], a_list[right_mark] = a_list[right_mark], a_list[first]
    return right_mark

def quick_sort_median3(a_list):
    quick_sort_median3_helper(a_list, 0, len(a_list) - 1)

def quick_sort_median3_helper(a_list, first, last):
    if first < last:
        split = partition_median3(a_list, first, last)
        quick_sort_median3_helper(a_list, first, split - 1)
        quick_sort_median3_helper(a_list, split + 1, last)

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
quick_sort_median3(lst)
print("Median-of-three pivot:", lst)
````
